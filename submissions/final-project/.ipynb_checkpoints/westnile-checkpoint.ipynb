```python
#Create a heat map of the occurrence of West Nile virus in the Chicago area.

from sklearn.neighbors import KernelDensity

chicago_map = np.loadtxt("mapdata_copyright_openstreetmap_contributors.txt")
#Create new dataframe
samples = pd.read_csv('train.csv')[['Date', 'Trap','Longitude', 'Latitude', 'WnvPresent']]

heat_plot = plt.cm.Reds
heat_plot._init()
heat_plot._lut[:-3,-1] = abs(np.logspace(0, 1, heat_plot.N) / 10 - 1)[::-1]
aspect = chicago_map.shape[0] * 1.0 / chicago_map.shape[1]
longitude_latitude_box = (-88, -87.5, 41.6, 42.1)

#new dataframe for sightings
sightings = samples[samples['WnvPresent'] > 0]
sightings = sightings.groupby(['Date', 'Trap','Longitude', 'Latitude']).max()['WnvPresent'].reset_index()
X = sightings[['Longitude', 'Latitude']].values
kd = KernelDensity(bandwidth=0.02)
kd.fit(X)

#Plotting and vectorizing sightings and occurrence of the virus
x_vector,y_vector = np.meshgrid(np.linspace(-88, -87.5, 100), np.linspace(41.6, 42.1, 100))
gridpoints = np.array([x_vector.ravel(),y_vector.ravel()]).T
z_vector = np.exp(kd.score_samples(gridpoints).reshape(100,100))
plt.figure(figsize=(10,14))
plt.imshow(chicago_map, 
           cmap=plt.get_cmap('gray'), 
           extent=longitude_latitude_box, 
           aspect=aspect)
plt.imshow(z_vector, 
           origin='lower', 
           cmap=heat_plot, 
           extent=longitude_latitude_box, 
           aspect=aspect)

loc = samples[['Longitude', 'Latitude']].drop_duplicates().values
plt.scatter(loc[:,0], loc[:,1], marker='o')

plt.savefig('heatmap.jpg')
```

```python
#Attempt to evalute the clustering
from sklearn.cluster import KMeans

def evaluate_clustering(X, max_k):
    """Plots the inertia of a clustering for different choices of cluster numbers.
    
    Parameters
    ----------
    X : matrix
        The data to be clustered.
    max_k : int
        The maximum number of clusters to try.
        
    Returns
    -------
    inertia : list
        The inertia for each choice of k.
    """
    inertia = []
    inertia.append(0)
    for k in range(1, max_k + 1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=5)
        kmeans.fit_predict(X)
        inertia.append(kmeans.inertia_)
    return inertia
```

```python
from numpy import hstack
from numpy import atleast_2d, where

data = hstack((atleast_2d(scaledX).T, atleast_2d(scaledY).T))
inertia_k = evaluate_clustering(data, 15)
print inertia_k
```

```python
#Graph trap infection rate.
#establish species types. Create column
SPECIES = ['CULEX PIPIENS', 
          'CULEX PIPIENS/RESTUANS',
          'CULEX RESTUANS']

#read in data. Create new dataframe. Parse the date and get year and weeks.
#The result entails that the virus is prevalent in not only seasons, but yearly cycles as well. 
samples = pd.read_csv('train.csv', parse_dates=['Date'])[['Species', 'WnvPresent', 'Date', 'Trap']]
samples = samples.groupby(['Date', 'Trap', 'Species']).max().reset_index()

samples['Year'] = samples['Date'].map(lambda x: x.year)
samples['Week'] = samples['Date'].map(lambda x: x.week)

samples = samples[['Year','Week','WnvPresent','Species']]
checks = samples[['Week', 'Year','WnvPresent']].groupby(['Week','Year']).count().reset_index()

wkly_pos = samples.groupby(['Year','Week', 'Species']).sum().reset_index()
wkly_pos_species = wkly_pos.set_index(['Year','Week', 'Species']).unstack()
wkly_pos_species.columns = wkly_pos_species.columns.get_level_values(1)
wkly_pos_species['total_positives'] = wkly_pos_species.sum(axis=1)
wkly_pos_species = wkly_pos_species.reset_index().fillna(0)
    
wkly_checks = checks.groupby(['Year','Week']).sum()
wkly_checks.columns = ['checks']
wkly_checks = wkly_checks.reset_index()
wkly_checks['positive'] = wkly_pos_species['total_positives']
wkly_checks['trap_infection_rate'] = wkly_checks['positive'] / wkly_checks['checks'] * 100
wkly_checks_yrs = wkly_checks.pivot(index='Week', columns='Year', values='trap_infection_rate')

ax = wkly_checks_yrs.interpolate().plot(title='Trap Infection Rate', figsize=(10,6))
ax.set_ylabel('Percentage samples infected');
plt.savefig('pos_trap_rate.jpg')

#Investigation on whether mosquito species have any effect of probability carrying the virus
fig, axes = plt.subplots(2, 2, figsize=(16,10), sharey=True, sharex=True)
fig.tight_layout()
axit = (ax for ax in it.chain(*axes))

#graphing percentage of mosquitoes(based on species) tested positive in the traps for each year.
for m, group in wkly_pos_species.groupby('Year'):
    ax = next(axit); ax.xaxis.grid(); ax.yaxis.grid()
    for species in SPECIES:
        ax.plot(group['Week'], group[species], label=species)
    ax.legend(loc='upper left');
    ax.set_title(m)
plt.savefig('pos_samples.jpg')
```
