Let's make a sample request to check that everything works fine. We will retrieve the names of NYT best-seller lists.

````python
import requests

response = requests.get("http://api.nytimes.com/svc/books/v3/lists/names.json?api-key=%s"%(key))
print response
````

Before we generate the request string, we need to read the guidelines from [here](http://developer.nytimes.com/docs/books_api/Books_API_Best_Sellers#h2-requests). According to the documentation under the _BEST-SELLER LIST NAMES_ section, the request must follow this URI structure:
```
http://api.nytimes.com/svc/books/{version}/lists/names[.response_format]?api-key={your-API-key}
```
We need to replace *{version}* with *v3*, the *response_format* with *json* and include our secrete API key.

We can try to print the response that NYT returns.

````python
print response.text
````

It is not possible to read the raw response. Instead, we need to decode the raw response as JSON and use the `json` library to print it.

````python
import json
print json.dumps(response.json(), indent=2)
````

Now, this is much better! We can easily see that the response consists of a response status, the number of results and a list of the best-seller lists. For each of these lists, we get information about its name, its update frequency, its lifetime and its codename.

## Step 2. Parsing the responses

In this section, we practice some of the basic Python tools that we learned in class and the powerful string handling methods that Python offers. Our goal is to be able to pick the interesting parts of the response and transform them in a format that will be useful to us.

Our first task will be to isolate the names of all the best-seller lists of the NYT. Fill in the rest of the ```print_names_from_XML()``` function that reads the XML response and prints all these names. **(5 pts)** 

Hint: Our _pretty_ formatter puts each tag on a separate line. You may want to read the documentation of the `split()`, `strip()` and `startswith()` functions.

````python
import re
def print_names_from_XML(response):
    xml_parser = xml.dom.minidom.parseString(response.text)
    pretty_response = xml_parser.toprettyxml()
    
    #Set the start and end conditions to determine the title names
    start= r'<list_name>'
    end= r'</list_name>'
    
    #Find the index values for the start and end of the list
    startTitles = [x.start() for x in re.finditer(start, pretty_response)]
    endTitles = [x.end() for x in re.finditer(end, pretty_response)]
    
    #print the values between the 2 index values (start and end points)
    numTitles = len(startTitles)
    for i in range (1, numTitles):
        startIndex= startTitles[i]+len(start)
        endIndex = endTitles[i]-len(end)
        print pretty_response[startIndex:endIndex]
    return 
````

````python
response = requests.get("http://api.nytimes.com/svc/books/v3/lists/names.xml?api-key=%s"%(key))
print_names_from_XML(response)
````

Can you do the same thing for the JSON response? Notice that a JSON object is basically a dictionary. **(5 pts)**

````python
def print_names_from_JSON(response):
    """Prints the names of all the best-seller lists that are in the response.
    
    Parameters:
        response: Response object
        The response object that is a result of a get request for the names of the
        best-selling lists from the Books API. 
    
    """     
    #uses the same method after formatting using json
    pretty_response = json.dumps(response.json())
    
    #print pretty_response in JSON response
    start= r'display_name": "'
    end= r'", "list_name_encoded"'
    
    startTitles = [x.start() for x in re.finditer(start, pretty_response)]
    endTitles= [x.end() for x in re.finditer(end, pretty_response)]
    
    numTitles = len(startTitles)
    
    for i in range (1, numTitles):
        startIndex= startTitles[i]+len(start)
        endIndex = endTitles[i]-len(end)
        print pretty_response[startIndex:endIndex]
    return 
````

````python
response = requests.get("http://api.nytimes.com/svc/books/v3/lists/names.json?api-key=%s"%(key))
print_names_from_JSON(response)
````

## Step 3. Putting it all together

We are now ready to tackle our original problem; to compile a summary for the best-sellers over a period of 2 months. 

First, we need to become confident working with dates. Since we want to issue requests that span a period of 2 months, we need to be able to automatically advance a day, without needing to keep the logistics of how many days each month has. To this end, we will use the `datetime` library. Here is an example

Write a function that, given a list name and a date, returns a tuple with the books that were best-sellers for that date and the date on which the list was published by the NYT. **(40 pts)**

````python
import datetime
import time

def get_books(date, list_name):
    books = []
    publishedDate = []
    
    response = requests.get("http://api.nytimes.com/svc/books/v3/lists/%s/%s.json?&api-key=%s"%(date,list_name,key))
    pretty_response = json.dumps(response.json(), indent=2)
    
    #Gets the number value after using json dump and formatting the file
    startNum = r'"num_results": '
    endNum = r'"results":'
    startResults = [x.start() for x in re.finditer(startNum, pretty_response)]
    endResults = [x.end() for x in re.finditer(endNum, pretty_response)]
    results = pretty_response[startResults[0]+len(startNum):endResults[0]-len(endNum)-5]
    results = int(results)
    print "Number of Results: " + str(results)
    
    #Set pretty_response back to an empty string. 
    #Stores pretty_response with the listed information offsetting by 20 titles at a time. 
    pretty_response = ""
    resultsNumIdx = int(results)
    for i in range (0,resultsNumIdx,20):
        offset = i
        response = requests.get("http://api.nytimes.com/svc/books/v3/lists/%s/%s.json?&offset=%s&api-key=%s"%(date,list_name,offset,key))
        pretty_response = pretty_response + json.dumps(response.json(), indent=2)
    
    #Find the title index of each entry and append it to the books list
    startTitle = r'"title": '
    endTitle = r'"contributor_note"'
    startTitleList = [x.start() for x in re.finditer(startTitle, pretty_response)]
    endTitleList = [x.end() for x in re.finditer(endTitle, pretty_response)]
    
    numStart = len(startTitleList)
    for i in range (0, numStart):
        startIdx = startTitleList[i]+len(startTitle)+1
        endIdx = endTitleList[i]-len(endTitle)-12
        books.append(pretty_response[startIdx:endIdx])
    
    #Find the best seller date of each entry and append it to the booksellers_date list
    start = r'"bestsellers_date"'
    end = r'"books"'
    startList = [x.start() for x in re.finditer(start, pretty_response)]
    endList = [x.end() for x in re.finditer(end, pretty_response)]
    startIndex = startList[0]+len(start)+3
    endIndex = endList[0]-len(end)-8
    publishedDate.append(pretty_response[startIndex:endIndex])

    return (books,publishedDate)

date = datetime.date(2014,7,1)
list_name = "Trade Fiction Paperback"
get_books(date, list_name)
````

Notice that the free API key that we have has a limit of 8 QPS (queries per second). If we send multiple queries and pass this limit, we will get back an error instead of the answer. To avoid this situation, a naive way is, after each query, to wait $1/8=0.125$ seconds. The command for this is 
```python
time.sleep(0.125)
```

Let's now test our function:

Great! The final step is to write a function that takes a time window and a list name, and returns a dictionary with the books that were best-sellers as keys and the number of weeks that they were in the list as values. **(30 pts)**

````python
import datetime

def most_popular(start_date, end_date, list_name):
    books = []
    weeks_on_list = []
    date = start_date
    response = requests.get("http://api.nytimes.com/svc/books/v3/lists/%s/%s.json?weeks-on-list?&api-key=%s"%(date,list_name,key))
    pretty_response = json.dumps(response.json(), indent=2)
    
    #Gets the number of results from the json dump
    startIdx = r'"num_results": '
    endIdx = r'"results":'
    #Loops through to find all values that start with 'num_results' and ends with 'results'
    startResults = [x.start() for x in re.finditer(startIdx, pretty_response)]
    endResults = [x.end() for x in re.finditer(endIdx, pretty_response)]

    #Obtains the number of resulting titles
    startResultsIdx = startResults[0]+len(startIdx)
    endResultsIdx = endResults[0]-len(endIdx)-5
    num_results = pretty_response[startResultsIdx:endResultsIdx]
    printResults = int(num_results)
    print "Number of Results: " + str(printResults)
    
    #Sets pretty_response into an empty string.
    #Refill the string by re-inputting the titles and its respective details by offsets of 20 a page. 
    pretty_response = ""
    endRangeIdx = int(num_results)
    for i in range (0,endRangeIdx,20):
        offset = i
        response = requests.get("http://api.nytimes.com/svc/books/v3/lists/%s/%s.json?&offset=%s&api-key=%s"%(date,list_name,offset,key))
        pretty_response = pretty_response + json.dumps(response.json(), indent=2)
        
    #Find and add the title at each index to the books list
    startTitle = r'"title": '
    endTitle = r'"contributor_note"'
    startTitleList = [x.start() for x in re.finditer(startTitle, pretty_response)]
    endTitleList = [x.end() for x in re.finditer(endTitle, pretty_response)]
    
     #Find and add the weeks on list at each index to the weeks_on_list list
    startWeeks = r'"weeks_on_list": '
    endWeeks= r'"article_chapter_link":'
    startWeeksList = [x.start() for x in re.finditer(startWeeks, pretty_response)]
    endWeeksList = [x.end() for x in re.finditer(endWeeks, pretty_response)]
    
    #Add the books and their number of weeks on the list to their lists
    endRange = len(startTitleList)
    for i in range (0,endRange):
        startPrettyIdx = startTitleList[i]+len(startTitle)+1
        endPrettyIdx = endTitleList[i]-len(endTitle)-12
        
        #adding each book title to the books list after repopulating pretty_response
        #Retrieve from the indices (range)
        books.append(pretty_response[startPrettyIdx: endPrettyIdx])
        
        startWeeksIdx = startWeeksList[i]+len(startWeeks)
        endWeeksIdx = endWeeksList[i]-len(endWeeks)-11
        #adding each title by weeks into the weeks list 
        weeks_on_list.append(pretty_response[startWeeksIdx:endWeeksIdx])
   
    #For the dictionary, add the book and its number of weeks on list     
    bookDict = dict([(books[0], 1), (books[1], 1)])   
    endRange = len(books)
    for i in range (0,endRange):
        bookStart = books[i]
        bookEnd = weeks_on_list[i]
        bookDict.update({bookStart:bookEnd})
    return bookDict

start_date = datetime.date(2014,6,1)
end_date = datetime.date(2014,8,31)
list_name = "Trade Fiction Paperback"

booksDict = most_popular(start_date, end_date, list_name)
for book in booksDict:
    print book, ":", booksDict[book]
````

Again, let's test our function. It might take a while to run (because of the QPS limit).
